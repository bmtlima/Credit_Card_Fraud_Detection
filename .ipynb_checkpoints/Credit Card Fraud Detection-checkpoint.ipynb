```python
def frst_dig(pop_data):
    if int(str(pop_data)[0]) != 0:
        return int(str(pop_data)[0])

first_fraud = []

for i in range(0, len(fraud)):
    to_add = frst_dig(fraud.iloc[i]['Amount'])
    if to_add != None:
        first_fraud.append(to_add)

ocur = {}
for el in first_fraud:
    ocur[el] = ocur.get(el, 0) + 1
    
benford_fraud = [ocur[i]/len(first_fraud) for i in range(1, 10)]

print(benford_fraud)
```

# 4. Modelling - Undersampling

Undersampling is a common technique used to deal with highly imbalanced datasets. It consists of reducing the size of the abundant class (in our case, legit transactions) and it is preferred when there's a sufficient amount of rare data available (fraudulent transactions). 

```python
fraud.describe()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_df = pd.concat([legit_sample, fraud], axis=0)

new_df
```

```python
def ml_model(model,new_df):
    X = df.iloc[:,:-1]
    y = df.iloc[:,-1]
 
    #Implementing cross validation
 
    k = 5
    kf = KFold(n_splits=k, random_state=None)
    
    acc_score = []
 
    for train_index , test_index in kf.split(X):
        X_train , X_test = X.iloc[train_index,:],X.iloc[test_index,:]
        y_train , y_test = y[train_index] , y[test_index]
     
        model.fit(X_train,y_train)
        pred_values = model.predict(X_test)
     
        acc = accuracy_score(pred_values , y_test)
        acc_score.append(acc)
     
    return acc_score
```

```python
# These will be used for visualization later

cv_mean = []
cv_std = []
```

```python
acc_log = ml_model(LogisticRegression(), new_df)
avg_log = sum(acc_log)/5

cv_mean.append(avg_log)
cv_std.append(np.std(acc_log))

print('Accuracy of each fold - {}'.format(acc_log))
print('Avg accuracy : {}'.format(avg_log))
```

```python
acc_svc = ml_model(SVC(), new_df)
avg_svc = sum(acc_svc)/5

cv_mean.append(avg_svc)
cv_std.append(np.std(acc_svc))

print('Accuracy of each fold - {}'.format(acc_svc))
print('Avg accuracy : {}'.format(avg_svc))
```

We could've implemented more ML algorithms, but a key aspect of fraud detection is that it needs to be quick, as customers shouldn't have to wait too long to make a credit card transaction. Because of this, although SVC provided a better average accuracy compared to Logistic Regression, we should take the latter because of its higher efficiency.
